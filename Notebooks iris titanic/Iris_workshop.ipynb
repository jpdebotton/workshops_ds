```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
iris = pd.read_csv("Iris.csv")
```

```python
iris["species"].unique()
```

```python
plt.scatter(x=iris["sepal_length"], y =iris["sepal_width"])
plt.xlabel("Sepal Lenght cm")
plt.ylabel("Sepal Width cm")

plt.show()
```

```python
plt.figure(figsize=(7,4))

p1 = plt.scatter(x = iris[iris["species"]=="setosa"]["sepal_length"], y  = iris[iris["species"]=="setosa"]["sepal_width"], color = "Teal", label = "Iris-setosa")
p2 = plt.scatter(x = iris[iris["species"]=="versicolor"]["sepal_length"], y  = iris[iris["species"]=="versicolor"]["sepal_width"], color = "maroon", label = "Iris-versicolor")
p3 = plt.scatter(x = iris[iris["species"]=="virginica"]["sepal_length"], y  = iris[iris["species"]=="virginica"]["sepal_width"], color = "olive", label = "Iris-virginica")

plt.xlabel("Sepal Length cm")
plt.ylabel("Sepal Width cm")
plt.legend()
#p2.axes.legend("Iris-versicolor")
#p3.axes.legend("Iris-virginica")
plt.show()
```
