```python
DataSet2['Tokens3']= DataSet2['tweetText'].str.lower()
DataSet2.head()
```

```python
from string import punctuation
import string
```

```python
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation,'')
    return text
```
